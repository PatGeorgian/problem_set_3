```python
# Re-importing necessary libraries after the environment reset
import matplotlib.pyplot as plt
import numpy as np

# Sample data for marriages and divorces per capita for specific years
years = [1900, 1950, 2000]
marriages_per_capita = [9.3, 11.0, 8.2]  # Example data
divorces_per_capita = [0.5, 2.5, 4.0]    # Example data

# Bar chart positions
x = np.arange(len(years))

# Create the bar chart
plt.figure(figsize=(8, 6))
bar_width = 0.4
plt.bar(x - bar_width / 2, marriages_per_capita, width=bar_width, label='Marriages per Capita', color='blue')
plt.bar(x + bar_width / 2, divorces_per_capita, width=bar_width, label='Divorces per Capita', color='red')

# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Per Capita', fontsize=12)
plt.title('Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)', fontsize=14)
plt.xticks(x, years)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
